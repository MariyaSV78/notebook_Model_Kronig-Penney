**General objective:** is to understand how a periodic potential gives rise to allowed and forbidden energy bands for an electron in a crystal.

The Kronig-Penney model, developed by Ralph Kronig and William Penney in 1931, is a simple, idealized model of a quantum mechanical system consisting of an infinite series of quantum wells, all of the same size $a$, separated by rectangular potential barriers of width $b$ and height $V_0$, with each "cell" (well + one barrier) having a length $d$, where $d=a+b$.

We consider an electron in a crystal modeled by a one-dimensional periodic potential of the Kronig–Penney type:
- The potential consists of barriers of height $V_0$=1.5 eV and width $c$ = 0.3 nm.
- Between the barriers, the electron moves freely over a distance $b$ = 0.6 nm.
- The period of the potential is therefore $a = b+c$ = 0.9 nm.

We choose the origin such that:
- Free region: $V(x) = 0$, for $x ∈ [0,b[$.
- Barrier region: $V(x) = V_0$, for $x ∈ [-c,0[$.

We first consider the case where the energy $E$ of the electron’s eigenstate is less than $V_0$.

1. Write the Schrödinger equation in the two regions $[−c,0[$ and $[0,b[$.

The time-independent Schrödinger equation is

$$
- \frac{\hbar^2}{2m} \frac{d^2 \psi(x)}{dx^2} + V(x)\psi(x) = E\psi(x)
$$

For the two regions:

- **Region I** $[-c, 0[$ (barrier region, $V(x) = V_0$):

  $$
  -\frac{\hbar^2}{2m} \frac{d^2 \psi_I(x)}{dx^2} + V_0\psi_I(x) = E\psi_I(x)
  \quad \Longrightarrow \quad
  \frac{d^2 \psi_I(x)}{dx^2} = \frac{2m(V_0 - E)}{\hbar^2}\psi_I(x)
  $$

- **Region II** $[0, b[$ (free region, $V(x) = 0$):

  $$
  -\frac{\hbar^2}{2m} \frac{d^2 \psi_{II}(x)}{dx^2} = E\psi_{II}(x)
  \quad \Longrightarrow \quad
  \frac{d^2 \psi_{II}(x)}{dx^2} = -\frac{2mE}{\hbar^2}\psi_{II}(x)
  $$



With the standard notation

$$
\beta = \sqrt{\frac{2m(V_0 - E)}{\hbar^2}},
\qquad
\alpha = \sqrt{\frac{2mE}{\hbar^2}},
$$

we can write:

$$
\psi_I''(x) = \beta^2 \psi_I(x) \quad \text{for } x \in [-c, 0[.
\qquad
\psi_{II}''(x) = -\alpha^2 \psi_{II}(x) \quad \text{for } x \in [0, b[.
$$

2. Give the general forms of the solutions in each region.

From the differential equations obtained previously the general solutions can be written as follows:

- **Region I** $[-c, 0[$ (barrier region, $V(x) = V_0$):

  Since the equation involves a positive exponent $\beta^2$,
  the solution is a combination of exponential functions:
  $$
  \psi_I(x) = A e^{\beta x} + B e^{-\beta x},
  $$
  where $A$ and $B$ are constants determined by boundary conditions.

- **Region II** $[0, b[$ (free region, $V(x) = 0$):

  Since the equation involves a negative exponent $-\alpha^2$,
  the solution is a combination of sine and cosine (or equivalently complex exponentials):
  $$
  \psi_{II}(x) = C e^{i\alpha x} + D e^{-i\alpha x},
  $$
  where $C$ and $D$ are constants determined by boundary conditions.

3. Write the continuity conditions that must be imposed on $\psi(x)$ and $\psi '(x)$ at $x = 0$.
- **Region I** $[-c, 0[$ (barrier region, $V(x) = V_0$):
$$
\psi_I(x)=A e^{\beta x}+B e^{-\beta x}, 
\qquad
\psi_I'(x)=\beta\bigl(A e^{\beta x}-B e^{-\beta x}\bigr),
$$

- **Region II** $[0, b[$ (free region, $V(x) = 0$):
$$
\psi_{II}(x)=C e^{i\alpha x}+D e^{-i\alpha x},
\qquad
\psi_{II}'(x)=i\alpha\bigl(C e^{i\alpha x}-D e^{-i\alpha x}\bigr).
$$

**Continuity at $x = 0 $** gives
$$
\psi_I(0)=\psi_{II}(0) \quad\Longrightarrow\quad A + B = C + D,
$$

$$
\psi_I'(0)=\psi_{II}'(0) \quad\Longrightarrow\quad \beta(A - B)=i\alpha(C - D).
$$

Equivalently, as a linear system for $(A,B)$ in terms of $(C,D)$:
$$
\begin{pmatrix} 1 & 1 \\ \beta & -\beta \end{pmatrix}
\begin{pmatrix} A \\ B \end{pmatrix}
=
\begin{pmatrix} 1 & 1 \\ i\alpha & -i\alpha \end{pmatrix}
\begin{pmatrix} C \\ D \end{pmatrix}.
$$

4. Recall Bloch’s theorem. Deduce from it the expression of the wave function in the interval $[b,a[$.
---
**Bloch's theorem:**
For any wave function that satisfies the Schrödinger equation in a periodic potential, and for a translation by one lattice vector $a$, there exists  at least one vector $k$ such that:

$$
\psi(x+a)=e^{i k a}\,\psi(x),
$$
where $\psi$ is the wave function, $k$ is the crystal wave vector.

---

The barrier in the reference cell is $x\in[-c,0[$. Adding the period $a$ maps this barrier to the next cell
$$
[-c,0[ \xrightarrow{+a} [a-c,a[ = [b,a[.
$$
For $x\in[b,a[$ set $x-a\in[-c,0[$ and use Bloch's relation:
$$
\psi(x)=\psi((x-a)+a)=e^{i k a}\,\psi(x-a).
$$

Using the barrier solution in the reference cell,
$$
\psi_I(x-a)=A e^{\beta(x-a)}+B e^{-\beta(x-a)},
$$
we obtain solution for the wave function in the barrier of the next cell $x\in[b,a[$
$$
\boxed{\,\psi(x)=e^{i k a}\bigl[A e^{\beta(x-a)}+B e^{-\beta(x-a)}\bigr]\, }.
$$

5. Write the continuity conditions that must be imposed on $\psi(x)$ and $\psi '(x)$ at $x = b$.

- **Region II** $[0, b[$ (free region, of the first cell, $V(x) = 0$):
$$
\psi_{II}(x)=C e^{i\alpha x}+D e^{-i\alpha x},
\qquad
\psi_{II}'(x)=i\alpha\bigl(C e^{i\alpha x}-D e^{-i\alpha x}\bigr).
$$

- **Region III** $[b, a[$ (barrier region in the next cell, $V(x) = V_0$):
$$
\psi_{III}(x)=e^{i k a} \bigl( A e^{\beta (x - a)}+B e^{-\beta (x - a)}\bigr), 
\qquad
\psi_{III}'(x)=e^{i k a} \beta \bigl( A e^{\beta (x - a)}-B e^{-\beta (x - a)}\bigr).
$$

**Continuity conditions at $x=b$:**

Wave function continuity:
$$
\psi_{II}(b) = \psi_{III}(b) \quad \Longrightarrow \quad
C e^{i \alpha b} + D e^{-i \alpha b} = e^{i k a} \Bigl( A e^{\beta (b-a)} + B e^{-\beta (b-a)} \Bigr) = e^{i k a} \Bigl( A e^{\beta c} + B e^{-\beta c} \Bigr)
$$

Derivative continuity:
$$
\psi_{II}'(b) = \psi_{III}'(b) \quad \Longrightarrow \quad
i \alpha \big( C e^{i \alpha b} - D e^{-i \alpha b} \big) = e^{i k a} \beta \big( A e^{\beta (b-a)} - B e^{-\beta (b-a)} \big) = e^{i k a} \beta \big( A e^{\beta c} - B e^{-\beta c} \big)
$$

Note:
- $C,D$ are the coefficients in the free region of the first cell (Region II).
- $A,B$ are the coefficients in the barrier of the next cell (Region III).
- $e^{i k a}$ comes from Bloch’s theorem, connecting the first cell to the next.

Together with the continuity at $x=0$ (Region I/II boundary), these conditions allow derivation of the dispersion relation.

6. Show that the dispersion relation can be written, in the case $ 0 < E < V_0 $:

Using the continuity conditions of the wave function and its derivative at $x=0$ and $x=b$ (from (1)- (3)):

$$
\psi_I(0)=\psi_{II}(0) \quad\Longrightarrow\quad A + B = C + D,
$$

$$
\psi_I'(0)=\psi_{II}'(0) \quad\Longrightarrow\quad \beta(A - B)=i\alpha(C - D).
$$

$$
\psi_{II}(b) = \psi_{III}(b) \quad \Longrightarrow \quad
C e^{i \alpha b} + D e^{-i \alpha b} = e^{i k a} \Bigl( A e^{\beta c} + B e^{-\beta c} \Bigr),
$$

$$
\psi_{II}'(b) = \psi_{III}'(b) \quad \Longrightarrow \quad
i \alpha \big( C e^{i \alpha b} - D e^{-i \alpha b} \big) = e^{i k a} \beta \big( A e^{\beta c} - B e^{-\beta c} \big).
$$

From the continuity at $x=0$ express $C$ and $D$ in terms of $A$ and $B$:

$$
C = \frac{1}{2} \Big( (A+B) - i \frac{\beta}{\alpha} (A-B) \Big), \qquad
D = \frac{1}{2} \Big( (A+B) + i \frac{\beta}{\alpha} (A-B) \Big).
$$

Substitute these into the expressions at $x=b$. Compute

$$
C e^{i \alpha b} + D e^{-i \alpha b}
= \frac{1}{2} \Big( (A+B)(e^{i \alpha b} + e^{-i \alpha b}) - i \frac{\beta}{\alpha} (A-B)(e^{i \alpha b} - e^{-i \alpha b}) \Big).
$$

Use the trigonometric identities

$$
e^{i \alpha b} + e^{-i \alpha b} = 2 \cos(\alpha b), \qquad 
e^{i \alpha b} - e^{-i \alpha b} = 2 i \sin(\alpha b),
$$

and the hyperbolic identities

$$
e^{\beta c} + e^{-\beta c} = 2 \cosh(\beta c), \qquad 
e^{\beta c} - e^{-\beta c} = 2 \sinh(\beta c).
$$

Then one obtains

$$
C e^{i \alpha b} + D e^{-i \alpha b} = (A+B)\cos(\alpha b) + \frac{\beta}{\alpha} (A-B) \sin(\alpha b),
$$

and

$$
i \alpha (C e^{i \alpha b} - D e^{-i \alpha b}) 
= (A+B) \alpha \sin(\alpha b) - (A-B) \beta \cos(\alpha b).
$$


Hence the homogeneous linear system  at $x=b$ becomes

$$
(A+B) \cos(\alpha b) + \frac{\beta}{\alpha} (A-B) \sin(\alpha b) = e^{i k a} \big( A e^{\beta c} + B e^{-\beta c} \big),
$$

$$
(A+B) \alpha \sin(\alpha b) - (A-B) \beta \cos(\alpha b) = e^{i k a} \beta \big( A e^{\beta c} - B e^{-\beta c} \big).
$$
---

With introduce the abbreviations
$$
X=e^{ika},\qquad u=A+B,\qquad v=A-B.
$$

$$
\begin{pmatrix}
\cos(\alpha b)-X\cosh(\beta c) & \dfrac{\beta}{\alpha}\sin(\alpha b)-X\sinh(\beta c)\\[8pt]
\alpha\sin(\alpha b)-X\beta\sinh(\beta c) & -\beta\cos(\alpha b)+X\beta\cosh(\beta c)
\end{pmatrix}
\begin{pmatrix} u\\ v\end{pmatrix}
= \begin{pmatrix}0\\0\end{pmatrix}.
$$

For a nontrivial solution, the determinant must vanish:

$$
\begin{vmatrix}
\cos(\alpha b)-X\cosh(\beta c) & \dfrac{\beta}{\alpha}\sin(\alpha b)-X\sinh(\beta c)\\[8pt]
\alpha\sin(\alpha b)-X\beta\sinh(\beta c) & -\beta\cos(\alpha b)+X\beta\cosh(\beta c)
\end{vmatrix} = 0.
$$

Expanding the determinant gives

$$
(\cos\alpha b - X\cosh\beta c)(-\beta\cos\alpha b + X\beta\cosh\beta c)
-
\left( \frac{\beta}{\alpha}\sin\alpha b - X\sinh\beta c \right)
\left( \alpha\sin\alpha b - X\beta\sinh\beta c \right) = 0.
$$

---

Simplify and group terms:

$$
-\beta\cos^2\!\alpha b + 2X\beta\cos\alpha b \cosh\beta c - X^2\beta\cosh^2\!\beta c
- \frac{\beta}{\alpha}\alpha\sin^2\!\alpha b + X\left(\frac{\beta^2}{\alpha}+\alpha\right)\sin\alpha b\sinh\beta c - X^2\beta\sinh^2\!\beta c = 0.
$$

$$
-\beta(\cos^2\!\alpha b + \sin^2\!\alpha b)
+ 2X\beta\cos\alpha b \cosh\beta c
- X^2\beta(\cosh^2\!\beta c + \sinh^2\!\beta c)
+ X\sin\alpha b\sinh\beta c\left( \frac{\beta^2}{\alpha} + \alpha \right) = 0.
$$

Since $\cos^2\alpha b + \sin^2\alpha b = 1$, we obtain

$$
-\beta + 2X\beta\cos\alpha b\cosh\beta c
- X^2\beta(\cosh^2\!\beta c + \sinh^2\!\beta c)
+ X\sin\alpha b\sinh\beta c\left( \frac{\beta^2}{\alpha} + \alpha \right) = 0.
$$

---

Divide by $\beta$ (nonzero)

$$
-1 + 2X\cos\alpha b\cosh\beta c
- X^2(\cosh^2\!\beta c + \sinh^2\!\beta c)
+ X\sin\alpha b\sinh\beta c\left( \frac{\beta}{\alpha} + \frac{\alpha}{\beta} \right) = 0.
$$

---
Rearrange as a quadratic in $X$

$$
(\cosh^2\!\beta c + \sinh^2\!\beta c) X^2
- \left[ 2\cos\alpha b\cosh\beta c + \left( \frac{\alpha}{\beta} + \frac{\beta}{\alpha} \right)\sin\alpha b\sinh\beta c \right] X
+ 1 = 0.
$$

Divide through by $X$ to obtain a relation involving $X + X^{-1}$:

$$
(\cosh^2\!\beta c + \sinh^2\!\beta c)\left(X + X^{-1}\right)
= 2\cos\alpha b\cosh\beta c + \left( \frac{\alpha}{\beta} + \frac{\beta}{\alpha} \right)\sin\alpha b\sinh\beta c.
$$

Using the identity $\cosh^2x + \sinh^2x = 2\cosh^2x - 1$, we can simplify further — after straightforward algebra, we obtain

$$
X + X^{-1} = 2\cos(\alpha b)\cosh(\beta c)
- \frac{\alpha^2 - \beta^2}{\alpha\beta}\sin(\alpha b)\sinh(\beta c).
$$

---
Express in terms of $\cos(ka)$

Since $X = e^{ika}$, it follows that

$$
X + X^{-1} = e^{ika} + e^{-ika} = 2\cos(ka).
$$

Hence, dividing both sides by $2$ gives

$$
\boxed{
\cos(ka) = \cos(\alpha b)\cosh(\beta c)
- \frac{\alpha^2 - \beta^2}{2\alpha\beta}\sin(\alpha b)\sinh(\beta c)
}
$$

with

$$
\alpha = \sqrt{\frac{2mE}{\hbar^2}}, \qquad
\beta = \sqrt{\frac{2m(V_0 - E)}{\hbar^2}}.

$$

---

This is the required dispersion relation for the case $0 < E < V_0$, describing the allowed energy bands in the Kronig–Penney model.

7. Dispersion relation in the case $V_0 < E$.

For $V_0<E$ the electron is oscillatory in both regions. Define
$$
\alpha=\sqrt{\frac{2mE}{\hbar^2}},\qquad
\kappa=\sqrt{\frac{2m(E-V_0)}{\hbar^2}}.
$$

**Region I (barrier in the reference cell), $x\in[-c,0[$**, now oscillatory:
$$
\psi_I(x)=A e^{i\kappa x}+B e^{-i\kappa x}.
$$

**Region II (free region), $x\in[0,b[$**:
$$
\psi_{II}(x)=C e^{i\alpha x}+D e^{-i\alpha x}.
$$

**Region III (barrier in the next cell, $x\in[b,a[$)** is obtained by Bloch's theorem:
$$
\psi_{III}(x)=e^{ika}\psi_I(x-a)=e^{ika}\big(A e^{i\kappa(x-a)}+B e^{-i\kappa(x-a)}\big).
$$

- Continuity and derivative continuity at $x=0$ give
$$
A+B=C+D,
$$
$$
i\kappa(A-B)=i\alpha(C-D).
$$
Hence
$$
C=\frac{1}{2}\Big((A+B)-\frac{\kappa}{\alpha}(A-B)\Big),\qquad
D=\frac{1}{2}\Big((A+B)+\frac{\kappa}{\alpha}(A-B)\Big).
$$

The factors of $i$ cancel because both sides contain $i$.

- Use the expressions for $C,D$ and evaluate at $x=b$. As before one obtains a homogeneous linear system for
$$
u=A+B,\qquad v=A-B,
$$
of the form
$$
\begin{pmatrix}
\cos(\alpha b)-X\cos(\kappa c) & \dfrac{\kappa}{\alpha}\sin(\alpha b)-X\sin(\kappa c)\\[8pt]
\alpha\sin(\alpha b)-X\kappa\sin(\kappa c) & -\kappa\cos(\alpha b)+X\kappa\cos(\kappa c)
\end{pmatrix}
\begin{pmatrix} u\\ v\end{pmatrix}=0,
$$
where $X=e^{ika}$ and we used the trigonometric identities
$$
e^{i\kappa c}+e^{-i\kappa c}=2\cos(\kappa c),\qquad
e^{i\kappa c}-e^{-i\kappa c}=2i\sin(\kappa c).
$$


A non trivial solution exists if the determinant vanishes. The determinant expansion and algebra follow exactly the same pattern as in the $0<E<V_0$ case, except that every occurrence of $\beta$ and hyperbolic functions is replaced by $i\kappa$ and trigonometric functions. Equivalently, perform the substitution
$$
\beta\mapsto i\kappa,\qquad \cosh(\beta c)\mapsto\cos(\kappa c),\qquad \sinh(\beta c)\mapsto i\sin(\kappa c)
$$
in the dispersion relation of the $0<E<V_0$ case:
$$
\cos(ka)=\cos(\alpha b)\cosh(\beta c)
-\frac{\alpha^2-\beta^2}{2\alpha\beta}\sin(\alpha b)\sinh(\beta c).
$$

Using $\beta=i\kappa$ we get
$$
\cosh(\beta c)=\cosh(i\kappa c)=\cos(\kappa c),\qquad
\sinh(\beta c)=\sinh(i\kappa c)=i\sin(\kappa c),
$$
and
$$
\frac{\alpha^2-\beta^2}{2\alpha\beta}
=\frac{\alpha^2-(i\kappa)^2}{2\alpha(i\kappa)}
=\frac{\alpha^2+\kappa^2}{2i\alpha\kappa}.
$$
Substituting into the previous formula and cancelling the factor $i$ coming from $\sinh(\beta c)$, one obtains the dispersion relation for $V_0<E$:
$$
\boxed{\;
\cos(ka)=\cos(\alpha b)\cos(\kappa c)
- \frac{\alpha^2+\kappa^2}{2\alpha\kappa}\sin(\alpha b)\sin(\kappa c)
\;}
$$
with
$$
\boxed{\;
\alpha=\sqrt{\frac{2mE}{\hbar^2}},\qquad
\kappa=\sqrt{\frac{2m(E-V_0)}{\hbar^2}}.
\;}
$$

8. Using a numerical tool (Python, MATLAB, etc.), plot $\cos(k a)$ as a function of $ E ∈ [0,10]$ eV.

9. Identify the energy intervals for which $∣cos(k a)∣ ≤ 1$. These correspond to the allowed bands.

```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import groupby
from operator import itemgetter

# Constants
hbar = 1.0545718e-34   # Reduced Planck constant [J·s]
m = 9.10938356e-31     # Electron mass [kg]
eV = 1.60218e-19       # 1 eV in Joules

# Kronig-Penney parameters
V0 = 5.0 * eV          # Barrier height [J]
b = 1e-10              # Width of free region [m]
c = 0.5e-10            # Width of barrier [m]
a = b + c              # Lattice period

# Energy range [0,10 eV]
E_eV = np.linspace(0.01, 10, 1000)  # Avoid E=0 for sqrt
E = E_eV * eV

# Initialize cos(ka)
cos_ka = np.zeros_like(E)

# Compute alpha, beta/kappa and cos(ka)
for i, Ei in enumerate(E):
    alpha = np.sqrt(2 * m * Ei) / hbar
    if Ei < V0:
        beta = np.sqrt(2 * m * (V0 - Ei)) / hbar
        cos_ka[i] = (np.cos(alpha * b) * np.cosh(beta * c) 
        - ((alpha**2 - beta**2)/(2 * alpha * beta)) * np.sin(alpha * b) * np.sinh(beta * c))
    elif Ei > V0:
        kappa = np.sqrt(2 * m * (Ei - V0)) / hbar
        cos_ka[i] = (np.cos(alpha * b) * np.cos(kappa * c) 
        - ((alpha**2 + kappa**2) / (2 * alpha * kappa)) * np.sin(alpha * b) * np.sin(kappa * c))
    else:  # Ei == V0, limit case
        cos_ka[i] = np.cos(alpha * b) - 0.5 * alpha * c * np.sin(alpha * b)


# Identify allowed bands (|cos(ka)| <= 1)
allowed = np.abs(cos_ka) <= 1
indices = np.where(allowed)[0]

intervals = []
for k, g in groupby(enumerate(indices), lambda ix: ix[0]-ix[1]):
    group = list(map(itemgetter(1), g))
    intervals.append((E_eV[group[0]], E_eV[group[-1]]))

print("Allowed energy bands (eV):")
for start, end in intervals:
    print(f"{start:.3f} - {end:.3f}")

# Plot
plt.figure(figsize=(8,5))
plt.plot(E_eV, cos_ka, label=r'$\cos(ka)$')
plt.axhline(1, color='r', linestyle='--')
plt.axhline(-1, color='r', linestyle='--')
plt.xlabel('Energy E [eV]')
plt.ylabel(r'$\cos(ka)$')
plt.title('$\cos(ka)$ vs E')
plt.legend()
plt.grid(True)
plt.show()
```

10. Plot $E(k)−⟨V⟩$ in the first half of the Brillouin zone: $k∈[0,\pi/a]$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
hbar = 1.0545718e-34   # Reduced Planck constant [J·s]
m = 9.10938356e-31     # Electron mass [kg]
eV = 1.60218e-19       # 1 eV in Joules

# Kronig-Penney parameters
V0 = 5.0 * eV          # Barrier height [J]
b = 1e-10              # Width of free region [m]
c = 0.5e-10            # Width of barrier [m]
a = b + c              # Lattice period

# Energy range
E_eV = np.linspace(0.01, 10, 5000)
E = E_eV * eV

# Initialize arrays
cos_ka = np.zeros_like(E)

# Compute cos(ka)
for i, Ei in enumerate(E):
    alpha = np.sqrt(2*m*Ei)/hbar
    if Ei < V0:
        beta = np.sqrt(2*m*(V0 - Ei))/hbar
        cos_ka[i] = (np.cos(alpha*b)*np.cosh(beta*c) 
                     - ((alpha**2 - beta**2)/(2*alpha*beta))*np.sin(alpha*b)*np.sinh(beta*c))
    elif Ei > V0:
        kappa = np.sqrt(2*m*(Ei - V0))/hbar
        cos_ka[i] = (np.cos(alpha*b)*np.cos(kappa*c) 
                     - ((alpha**2 + kappa**2)/(2*alpha*kappa))*np.sin(alpha*b)*np.sin(kappa*c))
    else:  # Ei == V0
        cos_ka[i] = np.cos(alpha*b) - 0.5*alpha*c*np.sin(alpha*b)

# Allowed bands: |cos(ka)| <= 1
allowed = np.abs(cos_ka) <= 1
E_allowed = E[allowed]
cos_ka_allowed = cos_ka[allowed]

# Compute k for allowed energies
k_allowed = np.arccos(cos_ka_allowed)/a  # First Brillouin zone: k ∈ [0, π/a]

# Average potential
V_avg = V0 * (c/a)

# Plot E(k) - <V>
plt.figure(figsize=(8,5))
plt.plot(k_allowed, E_allowed/eV - V_avg/eV, '.', markersize=1)
plt.axvline(np.pi/a, color='r', linestyle='--', label='First Brillouin zone edge (π/a)')

plt.xlabel('k [1/m]')
plt.ylabel('E(k) - <V> [eV]')
plt.title('Energy bands in the first Brillouin zone')
# plt.xlim(0, np.pi/a)
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
hbar = 1.0545718e-34   # Reduced Planck constant [J·s]
m = 9.10938356e-31     # Electron mass [kg]
eV = 1.60218e-19       # 1 eV in Joules

# Kronig-Penney parameters
V0 = 5.0 * eV          # Barrier height [J]
b = 1e-10              # Width of free region [m]
c = 0.5e-10            # Width of barrier [m]
a = b + c              # Lattice period

# Energy range
E_eV = np.linspace(0.01, 100, 5000)
E = E_eV * eV

# Compute cos(ka)
cos_ka = np.zeros_like(E)
for i, Ei in enumerate(E):
    alpha = np.sqrt(2*m*Ei)/hbar
    if Ei < V0:
        beta = np.sqrt(2*m*(V0 - Ei))/hbar
        cos_ka[i] = np.cos(alpha*b)*np.cosh(beta*c) - ((alpha**2 - beta**2)/(2*alpha*beta))*np.sin(alpha*b)*np.sinh(beta*c)
    elif Ei > V0:
        kappa = np.sqrt(2*m*(Ei - V0))/hbar
        cos_ka[i] = np.cos(alpha*b)*np.cos(kappa*c) - ((alpha**2 + kappa**2)/(2*alpha*kappa))*np.sin(alpha*b)*np.sin(kappa*c)
    else:  # Ei == V0
        cos_ka[i] = np.cos(alpha*b) - 0.5*alpha*c*np.sin(alpha*b)

# Allowed energies: |cos(ka)| <= 1
allowed = np.abs(cos_ka) <= 1
E_allowed = E[allowed]
cos_ka_allowed = cos_ka[allowed]

# Compute k for allowed energies in first Brillouin zone
k_allowed = np.arccos(cos_ka_allowed)/a

# Mirror for negative k (to show symmetry)
k_vals = np.concatenate([-k_allowed[::-1], k_allowed])
E_vals = np.concatenate([E_allowed[::-1], E_allowed])

# Extend to multiple zones
num_zones = 3  # Number of Brillouin zones
k_all = []
E_all = []

for n in range(-num_zones+1, num_zones):
    k_shifted = k_vals + 2*n*np.pi/a
    k_all.append(k_shifted)
    E_all.append(E_vals)

k_all = np.concatenate(k_all)
E_all = np.concatenate(E_all)

# Plot the energy bands
plt.figure(figsize=(10,6))
plt.plot(k_all, E_all/eV, '.', markersize=1, color='b')
for n in range(-num_zones, num_zones+1):
    plt.axvline(n*np.pi/a, color='r', linestyle='--', alpha=0.5)  # Brillouin zone boundaries
plt.xlabel('k [1/m]')
plt.ylabel('Energy [eV]')
plt.title('Kronig-Penney Model: Crystal Band Structure')
plt.grid(True)
plt.show()
```
